## üì•Ô∏è Install dependencies

We highly recommend to store the dependencies you install in files in a persistently saved folder (e.g. `env.yml`, `requirements.txt`) to reinstall everything easily in case the container is restarted. 

### üêç Install conda packages

You can use `conda` or `mamba` to install conda packages. Mamba is a reimplementation of conda in C++ that makes it much faster, every commands used in conda work in mamba, you just need to replace conda by mamba at the start of the command. 

For example to install plotly using mamba, you can run:


```bash
mamba install plotly
```


‚ö†Ô∏è If `mamba` does not work for you, then use `conda`

If you need to use a different Python version than the one used by default in the container you can create a new environment with the Python version you need. 

For example, to use an older version Python like 3.8 in a new conda environment named `py38`, your `env.yml` should look like this, you can add more conda and pip packages to install:


```yaml
name: py38
channels:
  - conda-forge
dependencies:
  - python=3.8
  - ipykernel 
  - pip
  - pip:
    - your-pip-package
```


Once your environment has been defined, you can create it:


```bash
mamba env create -f env.yml
```


‚ö†Ô∏è You cannot use `conda activate` in a Docker container, so you will need to create a new tile which will use the kernel from your conda environment. Running the following command will create the tile on the JupyterLab launcher page:


```bash
conda run -n py38 python -m ipykernel install --user --name py38 --display-name "Python 3.8 (py38)"
```


To verify you are using the right kernel you can run the following inside your notebook opened through the tile with using the kernel from your conda environment:

```python
import sys
!{sys.executable} --version
```

The output based on our example `env.yml` should be a version of Python 3.8: "Python 3.8.X"

Or use `conda run` to run scripts and/or commands directly using the new environment:


```bash
mamba run -n py38 python --version
```


### üêç Install pip packages

Install Python packages with `pip`:


```bash
pip install -r requirements.txt
```
